### FUTURE ENGINEERING AND DATA PROCESSING 

```python
# Reloading the cleaned dataset for futrue engineering 
loan_data = pd.read_csv('C:/csv_files/clean_loan_default.csv')
```

```python
# Display original data clreand data 
loan_data.head(3)
```

```python
# Future engineering and encode categorical
# 1. Convert Yes/No to Binary

binary_cols = ["HasMortgage", "HasDependents", "HasCoSigner"]
for col in binary_cols:
    loan_data[col] = loan_data[col].map({"Yes": 1, "No": 0})

# 2. Binning Numerical Features

# Age groups
loan_data["AgeGroup"] = pd.cut(loan_data["Age"], bins=[17, 25, 40, 60, 100],
                        labels=["18-25", "26-40", "41-60", "60+"] )

# Income bands
loan_data["IncomeBand"] = pd.cut(loan_data["Income"], bins=[0, 30000, 60000, 100000, float("inf")],
                          labels=["Low", "Medium", "Upper-Mid", "High"])

# Credit score bands
loan_data["CreditScoreBand"] = pd.cut(loan_data["CreditScore"], 
                               bins=[0, 580, 669, 739, 799, 850],
                               labels=["Poor", "Fair", "Good", "Very Good", "Excellent"])

# DTI Ratio bands
loan_data["DTIBand"] = pd.cut(loan_data["DTIRatio"], bins=[0, 0.3, 0.6, 1.0, float("inf")],
                       labels=["Low", "Medium", "High", "Extreme"])

# Months employed bins
loan_data["EmploymentStability"] = pd.cut(loan_data["MonthsEmployed"], bins=[-1, 12, 60, 120, float("inf")],
                                   labels=["<1yr", "1-5yrs", "5-10yrs", "10+yrs"])
# 3. Derived Features

# Loan-to-Income ratio
loan_data["LoanToIncomeRatio"] = loan_data["LoanAmount"] / loan_data["Income"]

# High risk flag
loan_data["HighRiskFlag"] = ((loan_data["CreditScore"] < 580) | (loan_data["DTIRatio"] > 0.6)).astype(int)

# 4. One-hot encode categorical variables

categorical_cols = ["Education", "EmploymentType", "MaritalStatus", "LoanPurpose",
                    "AgeGroup", "IncomeBand", "CreditScoreBand", "DTIBand", "EmploymentStability"]

#loan_data = pd.get_dummies(loan_data, columns=categorical_cols, drop_first=True)
```
